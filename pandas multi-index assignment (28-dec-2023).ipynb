#### 11) How many Smart TV were sold in Telangana & Karnataka during the year 2017?

```python
telangana = y.loc[('Telangana', slice(None), 'Smart TV'), ('2017', slice(None))]
karnataka = y.loc[('Karnataka', slice(None), 'Smart TV'), ('2017', slice(None))]
two_states = telangana.sum() + karnataka.sum()

print(telangana, '\n\n')
print(karnataka, '\n\n')

print(f'total no.of Smart TV\'s sold are : ', two_states.sum())
```
